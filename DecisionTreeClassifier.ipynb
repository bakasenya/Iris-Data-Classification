```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, recall_score, roc_auc_score, f1_score, matthews_corrcoef, confusion_matrix, classification_report
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv("data/Iris.csv")

print(data.info())

data = data.drop(["Id"], axis=1)

Q1 = data["SepalWidthCm"].quantile(0.25)
Q3 = data["SepalWidthCm"].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data[(data["SepalWidthCm"] < lower_bound) | (data["SepalWidthCm"] > upper_bound)]
print(f"Outliers: {outliers}")

data = data[(data["SepalWidthCm"] >= lower_bound) & (data["SepalWidthCm"] <= upper_bound)].copy()
data.shape
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
X = data.copy()
X = X.drop(columns=["Species"])
y = data["Species"]

y_encoded = encoder.fit_transform(y)

X_train , X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=9)
```

```python
model = DecisionTreeClassifier(max_depth=9, random_state=9)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)

print(f"Accuracy: {accuracy}")
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, X, y, cv=5)  # 5-fold cross-validation
print(f"Cross-Validation Scores: {scores}")
print(f"Average Cross-Validation Score: {scores.mean()}")
```

```python
# Display classification report for precision, recall, and F1-score
print(classification_report(y_test, y_pred))

# Confusion Matrix to see the number of true positives, false positives, etc.
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y_train_encoded = encoder.fit_transform(y_train)
y_test_encoded = encoder.transform(y_test)

feature1, feature2 = "SepalLengthCm", "SepalWidthCm"

X_train_selected = X_train[[feature1, feature2]]
X_test_selected = X_test[[feature1, feature2]]

model = DecisionTreeClassifier(max_depth=9, random_state=9)
model.fit(X_train_selected, y_train_encoded)

x_min, x_max = X_train_selected[feature1].min() - 1, X_train_selected[feature1].max() + 1
y_min, y_max = X_train_selected[feature2].min() - 1, X_train_selected[feature2].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                     np.linspace(y_min, y_max, 100))

grid_points = np.c_[xx.ravel(), yy.ravel()]
grid_df = pd.DataFrame(grid_points, columns=[feature1, feature2])

Z = model.predict(grid_df)
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10,12))
plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.Set1)
sns.scatterplot(x=X_test.iloc[:, 0], 
                y=X_test.iloc[:, 1], 
                hue=y_test_encoded, 
                edgecolor="k", 
                palette="Set1")
plt.xlabel(feature1)
plt.ylabel(feature2)
plt.title("DB of Decision Trees")
plt.show()
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(15, 8))  # Adjust the figure size
plot_tree(model, feature_names=X_train.columns, class_names=encoder.classes_, filled=True)
plt.show()
```

```python
feature_importance = model.feature_importances_
features = np.array(X_train.columns)

sns.barplot(x=feature_importance, y=features)
plt.xlabel("Importance Score")
plt.ylabel("Feature")
plt.title("Feature Importance in Decision Tree")
plt.show()
```
